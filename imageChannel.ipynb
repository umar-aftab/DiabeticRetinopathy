```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#imported for image segmentation
import os
from skimage import data,io
import cv2
```

```python
# This transfers all the images into 3 differnt channels and puts them into their respective folders
import os
dir = "RetinalImages/"
for file in os.listdir( dir ):
    if file.endswith( ".png" ):
        #print(file)
        fileop=os.path.join(dir+file)
        retina = io.imread(fileop)
        retina_red= retina.copy()
        retina_green=retina.copy()
        retina_blue=retina.copy()
        
        #green
        green = retina_green[:, :, (0,2)] =0
        retina_green[green]=[0, 255, 0]
        #plt.imshow(retina_green)
        cv2.imwrite('GreenImages/'+file,retina_green)
        
        #red
        red = retina_red[:, :, (1,2)] =0
        retina_red[red] = [255, 0, 0]
        #plt.imshow(retina_red)
        cv2.imwrite('RedImages/'+file,retina_red)
        
        #blue
        blue = retina_blue[:, :, (0,1)] =0
        retina_blue[blue]=[0, 0, 255]
        #plt.imshow(retina_blue)
        cv2.imwrite('BlueImages/'+file,retina_blue)
```

```python
file=os.path.join("hard-exudates.jpg")
retina = io.imread(file)
retina_red= retina.copy()
retina_green=retina.copy()
retina_blue=retina.copy()
io.imshow(retina)
io.show()
```

```python
green = retina_green[:, :, (0,2)] =0
retina_green[green]=[0, 255, 0]
plt.imshow(retina_green)
cv2.imwrite('greenImage.png',retina_green)
```

```python
red = retina_red[:, :, (1,2)] =0
retina_red[red] = [255, 0, 0]
plt.imshow(retina_red)
cv2.imwrite('redImage.png',retina_red)
```

```python
blue = retina_blue[:, :, (0,1)] =0
retina_blue[blue]=[0, 0, 255]
plt.imshow(retina_blue)
cv2.imwrite('blueImage.png',retina_blue)
```

```python
titles=['red','green','blue']
images=[red,green,blue]
fig = plt.figure()
n=1
for image,title in zip(images,titles):
    if image==red:
        a=fig.add_subplot(1,3,1)
        retina_red[red]=[255,0,0]
        plt.imshow(retina_red)
        a.set_title("RED")
    if image==green:
        a=fig.add_subplot(1,3,2)
        retina_green[green]=[0,255,0]
        plt.imshow(retina_green)
        a.set_title("GREEN")

    if image==blue:
        a=fig.add_subplot(1,3,3)
        retina_blue[blue]=[0,0,255]
        plt.imshow(retina_blue)
        a.set_title("BLUE")
    n+=1
fig.set_size_inches(np.array(fig.get_size_inches())*len(images))
plt.show()
```
